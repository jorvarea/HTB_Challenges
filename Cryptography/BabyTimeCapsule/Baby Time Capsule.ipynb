```python
# Step 6: Turn the number into a string of bytes that can be interpreted as text

def long_to_bytes(n):
    byte_array = bytearray()
    while n > 0:
        byte_array.append(n % 256)
        n //= 256
    return byte_array[::-1]

plaintext_bytes = long_to_bytes(plaintext_number)
print(plaintext_bytes)
```
